```python
#IMPORTING PYTHON LIBRARIES
import cv2
import numpy
```

```python
#TASK 4.1
#CREATING MY OWN IMAGE / CREATE AN IMAGE USING PYTHON
img = numpy.zeros((600,600,3))
cv2.imshow("pic" , img)
cv2.waitKey()
cv2.destroyAllWindows()

font = cv2.FONT_HERSHEY_TRIPLEX 
org = (120,200)
fontscale = 0.9
color = (0,255,0)
thickness = 1
image1 = cv2.putText(img, 'FLOPPY DISK BHAMMM', org, font, fontscale, color, thickness, cv2.LINE_AA)
cv2.imshow("photo", image1)
cv2.waitKey()
cv2.destroyAllWindows()

image2 = cv2.rectangle(img, (100, 100), (500,500), (0,255,165), 3)
cv2.imshow("pic" , image2)
cv2.waitKey()
cv2.destroyAllWindows()

image3 = cv2.rectangle(img, (100, 100), (500,570), (0,255,165), 3)
cv2.imshow("pic" , image2)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.circle(img, (300,350), 100, (0,165,255), 4)
cv2.imshow("pic" , image3)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
#TASK 4.2
#IMPORTING AND RESIZING PICTURES / CROP 2 IMAGES AND SWAP
photo1 = cv2.imread('zrnitsu.png')
photo1 = cv2.resize(photo1 , (720,720))
crop1 = photo1[115:450 , 120:450]
cv2.imshow("image", photo1)
cv2.waitKey()
cv2.destroyAllWindows()

photo2 = cv2.imread('nezuko.png')
photo2 = cv2.resize(photo2 , (720,720))
crop2 = photo1[115:450 , 120:450]
cv2.imshow("image", photo2)
cv2.waitKey()
cv2.destroyAllWindows()


photo3 = cv2.imread('tanjiro.jpg')
photo3 = cv2.resize(photo3 , (720,720))
crop3 = photo1[120:450 , 120:450]
cv2.imshow("image", photo3)
cv2.waitKey()
cv2.destroyAllWindows()

crop1 = numpy.copy(crop1)
crop2 = numpy.copy(crop2)
photo1[115:450 , 120:450]=crop2
photo2[115:450 , 120:450]=crop1


cv2.imshow("xyz",  photo1)
cv2.waitKey()
cv2.destroyAllWindows()

cv2.imshow("abc",  photo2)
cv2.waitKey()
cv2.destroyAllWindows()
```
